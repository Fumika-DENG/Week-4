```python
graph = {}
person_dic = {}
with open("pages_large.txt", 'r', encoding='utf-8') as f:
    datas = [tmp.split() for tmp in f.readlines()]
    # datas=[[1,shegu], [2,B], [3,C], ...]
    for tmp in datas:
        person_dic[tmp[1]] = tmp[0]
        graph[tmp[0]] = set()

# 保存所有关系
with open("links_large.txt", 'r', encoding='utf-8') as f:
    datas = [tmp.strip().split() for tmp in f.readlines()]
for tmp in datas:
    graph[tmp[0]].add(tmp[1])
    graph[tmp[1]].add(tmp[0])

    ##(delete)graph = {["lines[0]":[lines[1]]]}
        ##{"A":["B","C","D"],xxxx}の形のグラフ

from collections import deque

def find_shortest_path(start, goal):
    ##start = 0　間違えた。いらないです。
    visited = []
    queue = deque()
    prev = {}
    
    visited.append(start)
    queue.append(start)
    


    ##while queue.empty() != 0:    pythonにはempty()のgrammaがないので、直接while queue:か、while len(queue) != 0にする
    while queue:
        node = queue.popleft()

        if node == goal:
            path = []
            

            while node != start:    #Notice: It's not node != goal(it will be empty)
                path.append(node)
                node = prev[node]
            path.append(start)
            path.reverse()
            return path

        for child in graph[node]:
            if child not in visited:
                visited.append(child)
                prev[child] = node
                queue.append(child)
                ##node = queue.pop()　ここはいらないです。上の循環にあるので　　　重複しない要注意

    return "no path"

start = person_dic["渋谷"]
goal = person_dic["小野妹子"]
print(find_shortest_path(start, goal))

```
